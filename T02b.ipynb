```python
import torch
import numpy as np

from matplotlib import pyplot as plt
import matplotlib.image as mpimg
```

```python
img_path = './figs/1600px-Stadtbild_Mnchen.jpg'

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

img = mpimg.imread(img_path)
gray = rgb2gray(img) 
plt.imshow(gray, cmap="gray")
plt.show()
```

```python
M = torch.tensor(gray)
```

```python
U, S, V = M.svd(some=False)
```

```python
plt.plot(S)
plt.ylabel('Magnitude')
plt.title('Singular values')
plt.show()
```

```python
plt.loglog(S)
plt.ylabel('Magnitude')
plt.title('Singular values')
plt.show()
```
